# Normalize Values and Map to Desired Output Range

```python
import math

def scale(x):

    ampMin = np.amin(x)
    ampMax = np.amax(x)
      
    scaleFactor = 1 / (ampMax - ampMin) 
    
    # Return the normalized signal.
    return (x - ampMin) * (scaleFactor)
```

```python
normGlobalSalesActivision = scale(globalSalesActivision)
normGlobalSalesAtari = scale(globalSalesAtari)
normGlobalSalesBethesda = scale(globalSalesBethesda)
normGlobalSalesCapcom = scale(globalSalesCapcom)
normGlobalSalesEA = scale(globalSalesEA)
normGlobalSalesKonami = scale(globalSalesKonami)
normGlobalSalesMicrosoft = scale(globalSalesMicrosoft)
normGlobalSalesNamco = scale(globalSalesNamco)
normGlobalSalesNintendo = scale(globalSalesNintendo)
normGlobalSalesSega = scale(globalSalesSega)
normGlobalSalesSony = scale(globalSalesSony)
normGlobalSalesSquareEnix = scale(globalSalesSquareEnix)
normGlobalSalesTakeTwo = scale(globalSalesTakeTwo)
normGlobalSalesUbisoft = scale(globalSalesUbisoft)

normCriticScoreActivision = scale(criticScoreActivision)
normCriticScoreAtari = scale(criticScoreAtari)
normCriticScoreBethesda = scale(criticScoreBethesda)
normCriticScoreCapcom = scale(criticScoreCapcom)
normCriticScoreEA = scale(criticScoreEA)
normCriticScoreKonami = scale(criticScoreKonami)
normCriticScoreMicrosoft = scale(criticScoreMicrosoft)
normCriticScoreNamco = scale(criticScoreNamco)
normCriticScoreNintendo = scale(criticScoreNintendo)
normCriticScoreSega = scale(criticScoreSega)
normCriticScoreSony = scale(criticScoreSony)
normCriticScoreSquareEnix = scale(criticScoreSquareEnix)
normCriticScoreTakeTwo = scale(criticScoreTakeTwo)
normCriticScoreUbisoft = scale(criticScoreUbisoft)
```

```python
# map the vector to a desired output range.

def mapToRange(x, lowP, highP):

    pFactors = scale(x)
    
    return pFactors*(highP - lowP) + lowP
```

```python
# Global Sales Mapped
gs0 = mapToRange(globalSalesActivision, 1, 100)
gs1 = mapToRange(normGlobalSalesAtari, 1, 100)
gs2 = mapToRange(normGlobalSalesBethesda, 1, 100)
gs3 = mapToRange(normGlobalSalesCapcom, 1, 100)
gs4 = mapToRange(normGlobalSalesEA, 1, 100)
gs5 = mapToRange(normGlobalSalesKonami, 1, 100)
gs6 = mapToRange(normGlobalSalesMicrosoft, 1, 100)
gs7 = mapToRange(normGlobalSalesNamco, 1, 100)
gs8 = mapToRange(normGlobalSalesNintendo, 1, 100)
gs9 = mapToRange(normGlobalSalesSega, 1, 100)
gs10 = mapToRange(normGlobalSalesSony, 1, 100)
gs11 = mapToRange(normGlobalSalesSquareEnix, 1, 100)
gs12 = mapToRange(normGlobalSalesTakeTwo, 1, 100)
gs13 = mapToRange(normGlobalSalesUbisoft, 1, 100)
```

```python
# Average Critic Score Mapped
cs0 = mapToRange(criticScoreActivision, 1, 10)
cs1 = mapToRange(normCriticScoreAtari, 1, 10)
cs2 = mapToRange(normCriticScoreBethesda, 1, 10)
cs3 = mapToRange(normCriticScoreCapcom, 1, 10)
cs4 = mapToRange(normCriticScoreEA, 1, 10)
cs5 = mapToRange(normCriticScoreKonami, 1, 10)
cs6 = mapToRange(normCriticScoreMicrosoft, 1, 10)
cs7 = mapToRange(normCriticScoreNamco, 1, 10)
cs8 = mapToRange(normCriticScoreNintendo, 1, 10)
cs9 = mapToRange(normCriticScoreSega, 1, 10)
cs10 = mapToRange(normCriticScoreSony, 1, 10)
cs11 = mapToRange(normCriticScoreSquareEnix, 1, 10)
cs12 = mapToRange(normCriticScoreTakeTwo, 1, 10)
cs13 = mapToRange(normCriticScoreUbisoft, 1, 10)
```
